```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_country = pd.read_excel('Country-Code[1].xlsx')
```

```python
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country__values = final_df.Country.value_counts().values
country__values
```

```python
#Pie chart for top 3 countries
plt.pie(country__values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation:Zomato maximum records or transactions are from India.

```python
city_names=final_df.City.value_counts().index
city_names
```

```python
city__values = final_df.City.value_counts().values
city__values
```

```python
# top 5 cities 
plt.pie(city__values[:5],labels=city_names[:5],autopct='%1.2f%%')
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

Observation

1.   when rating is between 4.5 to 4.9 ---> Excellent
2.   when ratings are between 4.0 to 4.4 ---> very good
3.   when ratings are between 3.5 to 3.9 ---> good
4.   when ratings are between 2.5 to 3.4 ---> average
5.   when ratings are between 1.0 to 2.4 ---> poor


```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y="Rating Count",data=ratings)
```

```python
#palette for mapping desired colors
sns.barplot(x='Aggregate rating',y="Rating Count",hue = 'Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observation 
1. Not rated count is very high.
2. Max no of rating are between 2.5 to 3.4

```python
## Count plot--
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
# Country by 0 aggregate rating counts
final_df[final_df['Aggregate rating']==0].groupby('Country').size().reset_index()
```

```python
# Country and its currency
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```
